```julia
# Import the Packages
using PrettyTables
using Plots, Printf
using DelimitedFiles   # Utilities for reading and writing delimited files, for example ".csv". See readdlm and writedlm
using CSV
using NBInclude        # Calling a julia function from jupyter notebook
```

```julia
## booleans
show_video = true
anim_pendulum = Animation()

## Give the initialization terms and values
int_method = "central_diff"
delta_t = 1.0e-3
steps = 1000               # for 2000, 3000, and 4000 also
```

```julia
## print this line first when the run start
println("-- pendulum ", int_method, " --")
```

```julia
## load pendulum
#@nbinclude("Math_pendulumProjectA.ipynb")
#@nbinclude("IntegratorProjectA.ipynb")
@nbinclude("DynsysProjectA.ipynb")

# As we know k = {(m * g) / l} for a pendulam
# Given: phi (t=0) = 1.0, phi_dot = 0.0, k = 300.0, c = 5.0, m = 1.0, g = 10.0 ==> so, l = (1 * 10)/300 = 10/300
# Math_pendulum(l, g, m, c, phi, phi_dot)

pendulum = Dynsys.Math_pendulum(10/300, 10.0, 1.0, 5.0, 1.0, 0.0)
```

```julia
## load integrator and memory for the results
Integ = Dynsys.Integrator(delta_t,steps)
Integ.res_phi = zeros(Integ.timesteps)
Integ.res_phi_dot = zeros(Integ.timesteps)

## run time integration
# initial setting
Figure = Dynsys.create_fig(pendulum)
Dynsys.plot_state(pendulum)
display(Figure)

# running over the time step
for i in 1:Integ.timesteps
    # integration step
    step = i
    
    # plot the state
    Figure = Dynsys.create_fig(pendulum)
    Dynsys.run_step(Integ, int_method, pendulum, step)
    Dynsys.plot_state(pendulum)
    display(Figure)
    frame(anim_pendulum)
    
    # save the step
    Integ.res_phi[i] = pendulum.phi
    Integ.res_phi_dot[i] = pendulum.phi_dot
end
```

```julia
## plot coordinate phi and the time derivative phi_dot

# time
x = LinRange(1, Integ.timesteps, Integ.timesteps)
x = x * Integ.delta_t

# plot phi results
y1 = Integ.res_phi
Figure1 = plot(x, y1 , xlabel = "time (s)", ylabel = "\u03C6", label = "Exact solution", lw = 3, lc = "black", 
        xtickfontsize=14, xguidefontsize=14, ytickfontsize=14, yguidefontsize=14, legendfontsize=14)

plot!(size=(800,600))

display(Figure1)

# save the figure
savefig("Phi_CentralDifferenceExactSolution_1000Points.png")
savefig("Phi_CentralDifferenceExactSolution_1000Points.pdf")


# phi_dot symbol
phi_dot = 'φ' * '̇'
#println(phi_dot)

# plot phi_dot results
y2 = Integ.res_phi_dot
Figure2 = plot(x, y2 , xlabel = "time (s)", ylabel = phi_dot, label = "Exact solution", lw = 3, lc = "black", 
        xtickfontsize=14, xguidefontsize=14, ytickfontsize=14, yguidefontsize=14, legendfontsize=14)

plot!(size=(800,600))

display(Figure2)

# save the figure
savefig("PhiDot_CentralDifferenceExactSolution_1000Points.png")
savefig("PhiDot_CentralDifferenceExactSolution_1000Points.pdf")


# Show and Save the animated video
if show_video
    gif(anim_pendulum, "PendulumVideo_CentralDifference_1000Points.gif")
end

## write the results data in a file
open("PendulamResults_CentralDifference_1000Points.txt", "w") do io
    writedlm(io, [x y1 y2])
end
```
